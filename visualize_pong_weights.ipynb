```python
import torch
import pickle
import numpy as np
from matplotlib import pyplot as plt
from pg_pong_pytorch import PongAgent
```

```python
def show_weights(weights, cols=5, rows=5):
    images = weights.reshape((200, 80, 80))
    
    fig, axes = plt.subplots(rows, cols, figsize=(2 * cols, 2 * rows))

    for image, ax in zip(images, axes.ravel()):
        ax.imshow(image, cmap='gray')
        ax.set_xticks(())
        ax.set_yticks(())

    plt.show()
```

```python
with open('save.p', 'rb') as f:
    numpy_model = pickle.load(f)
```

```python
w1_numpy = numpy_model['W1']
```

```python
show_weights(w1_numpy)
```

```python
with open('model_pytorch.pkl', 'rb') as f:
    pytorch_model = pickle.load(f)
```

```python
w1_torch = list(pytorch_model.fc1.parameters())[0].data.numpy()
```

```python
show_weights(w1_torch)
```
